```python
import sagemaker
from sagemaker.model import Model
from sagemaker import get_execution_role

# Initialize SageMaker session
sess = sagemaker.Session()
role = get_execution_role()

# Replace this with your image URI (from your build output)
image_uri = "275057777746.dkr.ecr.us-east-2.amazonaws.com/rfm-segmentation:latest"

# Create a SageMaker model
model = Model(
    image_uri=image_uri,
    role=role,
    sagemaker_session=sess
)

# Deploy model as an endpoint
predictor = model.deploy(
    initial_instance_count=1,
    instance_type="ml.m5.large",
    endpoint_name="rfm-segmentation-endpoint"
)
```

```python
import boto3

sm = boto3.client("sagemaker")

endpoint_name = "rfm-segmentation-endpoint"

# Delete endpoint (if exists)
try:
    sm.delete_endpoint(EndpointName=endpoint_name)
    print(f"✅ Deleted endpoint: {endpoint_name}")
except sm.exceptions.ClientError:
    print("ℹ️ No existing endpoint found")

# Delete endpoint config
try:
    sm.delete_endpoint_config(EndpointConfigName=endpoint_name)
    print(f"✅ Deleted endpoint config: {endpoint_name}")
except sm.exceptions.ClientError:
    print("ℹ️ No existing endpoint config found")
```
